```python
import networkx as nx
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import LineString
import ast  # safer than eval

def load_graph(graph_fp):
    print(" Loading weighted graph...")
    G = nx.read_graphml(graph_fp)

    # Convert node IDs from strings to tuples
    mapping = {}
    for node in G.nodes:
        try:
            mapping[node] = ast.literal_eval(node)
        except Exception:
            pass
    G = nx.relabel_nodes(G, mapping)

    return G

def plot_graph(G, path_edges=None):
    print(" Plotting...")
    edge_lines = []
    edge_colors = []

    for u, v, data in G.edges(data=True):
        line = LineString([u, v])
        edge_lines.append(line)
        weight = float(data.get("weight", 1.0))
        edge_colors.append(weight)

    gdf = gpd.GeoDataFrame(geometry=edge_lines)
    gdf["weight"] = edge_colors

    ax = gdf.plot(
        figsize=(10, 10),
        column="weight",
        cmap="YlGnBu_r",
        linewidth=1.5,
        legend=True,
        legend_kwds={"label": "Coolness weight"},
    )

    if path_edges:
        path_lines = [LineString([u, v]) for u, v in path_edges]
        gdf_path = gpd.GeoDataFrame(geometry=path_lines)
        gdf_path.plot(ax=ax, color="red", linewidth=3, label="Shortest Path")

    plt.title("Coolness-weighted Road Network")
    plt.axis("off")
    plt.legend()
    plt.tight_layout()
    plt.show()

def main():
    graph_fp = "../data/routing/cool_graph.graphml"
    G = load_graph(graph_fp)

    # OPTIONAL: visualize shortest path if needed
    # You can also disable this block if you want only the network

    # Example source and destination (lat, lon) within your AOI
    source = (12.9565, 77.6400)
    target = (12.9640, 77.6410)

    try:
        path = nx.shortest_path(G, source=source, target=target, weight="weight")
        path_edges = list(zip(path[:-1], path[1:]))
    except Exception as e:
        print(" Path computation failed:", e)
        path_edges = None

    plot_graph(G, path_edges=path_edges)

if __name__ == "__main__":
    main()
```

```python
# scripts/08_test_routing_final.py

import networkx as nx
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import random
import ast

def load_graph(path):
    print(" Loading GraphML...")
    G = nx.read_graphml(path)

    try:
        mapping = {n: int(n) for n in G.nodes}
        G = nx.relabel_nodes(G, mapping)
    except:
        pass

    return G

def extract_linestring(geom_str):
    try:
        coords = ast.literal_eval(geom_str)
        return LineString(coords)
    except Exception:
        return None

def plot_path(G, path, title="Coolness-Weighted Path"):
    fig, ax = plt.subplots(figsize=(10, 10))
    path_edges = set(zip(path[:-1], path[1:]))

    for u, v, data in G.edges(data=True):
        geom = extract_linestring(data.get("geometry", "[]"))
        if not isinstance(geom, LineString):
            continue
        x, y = geom.xy
        color = 'red' if (u, v) in path_edges or (v, u) in path_edges else '#bbbbbb'
        linewidth = 2 if color == 'red' else 0.6
        ax.plot(x, y, color=color, linewidth=linewidth)

    ax.set_title(title)
    ax.set_axis_off()
    plt.tight_layout()
    plt.show()

def main():
    G = load_graph("../data/routing/cool_graph.graphml")

    print(" Cleaning graph...")
    G = nx.DiGraph(G)  # force directional for future routing, but test in undirected
    for u, v, data in G.edges(data=True):
        if 'weight' not in data or data['weight'] is None:
            data['weight'] = 1e6  # fallback high weight
        else:
            try:
                data['weight'] = float(data['weight'])
            except:
                data['weight'] = 1e6

    print(" Extracting largest connected component...")
    G_u = G.to_undirected()
    largest_cc = max(nx.connected_components(G_u), key=len)
    G_sub = G.subgraph(largest_cc).copy()
    print(f" Component has {len(G_sub.nodes)} nodes, {len(G_sub.edges)} edges")

    # Retry up to N times to find a valid path
    attempts = 10
    for i in range(attempts):
        src, dst = random.sample(list(G_sub.nodes), 2)
        try:
            print(f" Attempt {i+1}: Source={src}, Target={dst}")
            path = nx.shortest_path(G_sub, source=src, target=dst, weight='weight')
            print(f" Path found with {len(path)} nodes")
            plot_path(G_sub, path)
            return
        except nx.NetworkXNoPath:
            print(" No path found, retrying...")

    print(" Failed to find path in 10 attempts.")

if __name__ == "__main__":
    main()
```

```python
# scripts/09_debug_full_graph.py

import networkx as nx
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import ast

def extract_linestring(geom_str):
    try:
        coords = ast.literal_eval(geom_str)
        return LineString(coords)
    except Exception:
        return None

def main():
    G = nx.read_graphml("../data/routing/cool_graph.graphml")

    # Optional relabel to int
    try:
        mapping = {n: int(n) for n in G.nodes}
        G = nx.relabel_nodes(G, mapping)
    except:
        pass

    fig, ax = plt.subplots(figsize=(12, 12))

    for u, v, data in G.edges(data=True):
        geom = extract_linestring(data.get("geometry", "[]"))
        if not isinstance(geom, LineString):
            continue
        x, y = geom.xy
        ax.plot(x, y, color="steelblue", linewidth=1)

    ax.set_title(" Full Routing Graph  Debug View")
    ax.axis("off")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
